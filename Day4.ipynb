```python
import numpy as np

# split data into 2D array
l = np.array([list(row) for row in test.split("\n")])
l = np.array([list(row) for row in actual.split("\n")])

l
```

```python
height, width = l.shape
```

# PART 1

```python
total_rolls = 0

# iterate through each position
for y, row in enumerate(l):
    for x, val in enumerate(row):
        if val == "@":
            # calculate bounds of box
            min_x = max([x - 1, 0])
            max_x = min([abs(x + 2), width])
            min_y = max([y - 1, 0])
            max_y = min([abs(y + 2), height])
            grid = l[min_y:max_y, min_x:max_x]

            # get count of @ in adjacent boxes
            total_count = np.char.count(grid, "@").sum()

            # add to total if fewer than 4 adjacent boxes
            if total_count < 5:
                total_rolls += 1

                # print(x, y)
                # print(grid)
                # print(f"{total_count = }\n")


total_rolls
```

# PART 2

```python
total_rolls = 0
old_rolls = -1
next_l = l.copy()

# iterate until no more rolls can be accessed
while total_rolls > old_rolls:
    old_rolls = total_rolls
    l = next_l.copy()

    # iterate through each position
    for y, row in enumerate(l):
        for x, val in enumerate(row):
            if val == "@":
                # calculate bounds of box
                min_x = max([x - 1, 0])
                max_x = min([abs(x + 2), width])
                min_y = max([y - 1, 0])
                max_y = min([abs(y + 2), height])
                grid = l[min_y:max_y, min_x:max_x]

                # get count of @ in adjacent boxes
                total_count = np.char.count(grid, "@").sum()

                # add to total if fewer than 4 adjacent boxes
                if total_count <= 4:
                    total_rolls += 1

                    # update grid for next go round
                    next_l[y, x] = x

                    # print(x, y)
                    # print(grid)
                    # print(f"{total_count = }\n")


total_rolls
```
