# PART 1

```python
def check(n):
    n = str(n)
    if len(n) % 2:
        return False

    else:
        n1 = n[: int(len(n) / 2)]
        n2 = n[int(len(n) / 2) :]

        if n1 == n2:
            return True
        else:
            return False
```

```python
res = 0

for n in l:
    if check(n):
        res += int(n)

res
```

# PART 2

```python
def factors(n):
    out = []
    for i in range(1, n // 2 + 1):
        if not n / i % 1:
            out.append(i)

    return out
```

```python
def check(n):
    n = str(n)

    for p in factors(len(n)):
        parts = [n[i : i + p] for i in range(0, len(n), p)]

        if len(set(parts)) == 1:
            return True

    return False
```

```python
res = 0

for n in l:
    if check(n):
        res += int(n)

res
```
