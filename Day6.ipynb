# PART 1

```python
import re

import numpy as np

sums = np.array(
    [re.sub("\s+", ",", s.strip()).split(",") for s in actual.split("\n")]
).T

sums
```

```python
def do_sum(l):
    if l[-1] == "+":
        return np.sum(l[:-1].astype("float64"))
    elif l[-1] == "*":
        return np.prod(l[:-1].astype("float64"))
```

```python
tot = 0
for row in sums:
    tot += do_sum(row)

int(tot)
```
