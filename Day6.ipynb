# PART 1

```python
import re

import numpy as np

sums = np.array(
    [re.sub("\s+", ",", s.strip()).split(",") for s in actual.split("\n")]
).T

sums
```

```python
def do_sum(l):
    if l[-1] == "+":
        return np.sum(l[:-1].astype("float64"))
    elif l[-1] == "*":
        return np.prod(l[:-1].astype("float64"))
```

```python
tot = 0
for row in sums:
    tot += do_sum(row)

int(tot)
```

# PART 2

```python
l = [[j for j in i] for i in actual.split("\n")]
# l = [[j for j in i] for i in test.split("\n")]

l[-1]
```

```python
operators = [i for i in l[-1] if i != " "]

operators

l2 = ["" for _ in range(len(l[0]))]

for i in range(len(l[0])):
    for j in range(len(l) - 1):
        val = l[j][i]
        # print(f"{val = }\n{i = }\n{j = }\n")
        if val.isdigit():
            l2[i] += val

l2 = [int(i) if i.isdigit() else operators.pop(0) for i in l2] + operators

l2
```

```python
operation = ""
temp_res = 0
res = 0
for i in l2[::-1] + [" "]:
    if isinstance(i, int):
        if operation == "*":
            temp_res *= i
        elif operation == "+":
            temp_res += i
    else:
        operation = i
        res += temp_res
        temp_res = 0 if i == "+" else 1


res
```
