```python
test = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""
```

```python
# I hope you're happy Harvey
with open("inputs/Day7_input.txt", "r") as file:
    actual = file.read()
```

```python
l = [[y for y in x] for x in test.split("\n")]
l = [[y for y in x] for x in actual.split("\n")]

l
```

# PART 1

```python
tot = 0

for y in range(len(l) - 1):
    for x in range(len(l[0])):
        val = l[y][x]

        if (val == "S") | (val == "|"):
            if l[y + 1][x] == ".":
                l[y + 1][x] = "|"
            elif l[y + 1][x] == "^":
                l[y + 1][x - 1] = "|"
                l[y + 1][x + 1] = "|"
                tot += 1

[print(row) for row in l]

tot
```

# PART 2

```python
l = [""]

l[0] = [[y for y in x] for x in test.split("\n")]
# l[0] = [[y for y in x] for x in actual.split("\n")]

l

tot = 0

for y in range(len(l[0]) - 1):
    extra_l = []
    for x in range(len(l[0][0])):
        for n in range(len(l) - 1):
            [print(i) for j in l for i in j]
            print(f"n:{n} y:{y} x:{x}")
            val = l[n][y][x]

            if (val == "S") | (val == "|"):
                if l[n][y + 1][x] == ".":
                    l[n][y + 1][x] = "|"
                elif l[n][y + 1][x] == "^":
                    l[n][y + 1][x - 1] = "|"
                    extra_l.append(l[n])
                    l[n][y + 1][x + 1] = "|"
                    tot += 1
    l.append(extra_l)

[print(row) for row in l]

tot
```
