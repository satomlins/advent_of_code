```python
def make_d(l):
    l = [x.split(" ") for x in l.split("\n")]

    d = {x[0][:3]: x[1:] for x in l}
    return d
```

# PART 1

```python
d = make_d(test)
```

```python
def recursive_count(item):
    if item == "out":
        return 1

    count = 0

    for v in d[item]:
        count += recursive_count(v)

    return count
```

```python
cache = {}


def recursive_count_2(item, b):
    if (item, b) in cache:
        return cache[(item, b)]

    match item:
        case "out":
            if all(b):
                return 1
            else:
                return 0

        case "dac":
            b = (1, b[1])
        case "fft":
            b = (b[0], 1)

    total = 0

    for v in d[item]:
        total += recursive_count_2(v, b)

    cache[(item, b)] = total
    return total


recursive_count_2("svr", (0, 0))
```
