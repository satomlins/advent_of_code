```python
def check_freshness(n):
    for f in fresh:
        min_f, max_f = [int(i) for i in f.split("-")]
        if (n >= min_f) & (n <= max_f):
            # print(f"{n = }\n{min_f = }\n{max_f = }\n")
            return 1
    return 0
```

```python
fresh_count = 0
for i in inventory:
    fresh_count += check_freshness(i)
fresh_count
```

```python
def calc_total(b):
    new_b = []
    for i in range(len(b[0])):
        new_b.append((b[0][i], b[1][i]))

    new_b = set(new_b)
    return sum([i - j + 1 for j, i in new_b])
```

```python
bounds = [[int(x.split("-")[0]) for x in fresh], [int(x.split("-")[1]) for x in fresh]]

old_total = -1
curr_total = 0

while curr_total != old_total:
    old_total = curr_total
    for i in range(len(bounds[0])):
        for j in range(len(bounds[0])):
            if i - j:
                if (bounds[0][i] >= bounds[0][j]) & (bounds[0][i] <= bounds[1][j]):
                    bounds[0][i] = bounds[0][j]

                if (bounds[1][i] >= bounds[0][j]) & (bounds[1][i] <= bounds[1][j]):
                    bounds[1][i] = bounds[1][j]

    curr_total = calc_total(bounds)


curr_total
```
