```python
test = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""
```

```python
big_string = test
big_string = actual
```

```python
# format inputs
fresh = big_string.split("\n\n")[0].split("\n")
inventory = [int(n) for n in big_string.split("\n\n")[1].split("\n")]
```

# PART 1

```python
# checks if item is fresh
def check_freshness(n):
    # iterate through fresh bounds
    for f in fresh:
        # get bounds
        min_f, max_f = [int(i) for i in f.split("-")]

        # check if item falls within bounds
        if (n >= min_f) & (n <= max_f):
            return 1
    return 0
```

```python
# calc total number of fresh items
fresh_count = 0
for i in inventory:
    fresh_count += check_freshness(i)
fresh_count
```

# PART 2

```python
# calculate total fresh items from bounds
def calc_total(b):
    return sum([i - j + 1 for j, i in b])
```

```python
# format bounds
bounds = [[int(n) for n in f.split("-")] for f in fresh]

old_total = -1
curr_total = 0

# run until no more bounds are changing
while curr_total != old_total:
    old_total = curr_total

    # iterate through all bounds, and then again
    for i in range(len(bounds)):
        for j in range(len(bounds)):
            # if we aren't comparing the same index
            if i - j:
                # if lower bound falls within range of another set of bounds
                if (bounds[i][0] >= bounds[j][0]) & (bounds[i][0] <= bounds[j][1]):
                    # change lower bound to be lower bound of that set of bounds
                    bounds[i][0] = bounds[j][0]

                # if upper bound falls within range of another set of bounds
                if (bounds[i][1] >= bounds[j][0]) & (bounds[i][1] <= bounds[j][1]):
                    # change upper bound to be upper bound of that set of bounds
                    bounds[i][1] = bounds[j][1]

    # drop duplicates
    unique_bounds = set(tuple(sorted(sub)) for sub in bounds)

    # convert back to list
    bounds = [list(tup) for tup in unique_bounds]

    # calculate new total
    curr_total = calc_total(bounds)


curr_total
```
